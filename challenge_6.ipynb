```python
from scipy import stats
import matplotlib.pyplot as plt

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(20, 15))

# Create some fake data with a normal distribution
normal_dist = stats.norm(loc=0, scale=1)
fake_y = normal_dist.rvs(1000)  

# Plot the data from the normal distribution
ax = axs[0]
ax.hist(fake_y, bins=20)  # Let's try something in between
ax.set_xlabel('Some x values')
ax.set_ylabel('Some x values')
ax.set_title("Normal Distribution")

# Create some fake data with an exponential distribution
exponential_dist = stats.expon(loc=0, scale=1)
fake_y = exponential_dist.rvs(1000)  

# Plot the data from the exponential distribution
ax = axs[1]
ax.hist(-fake_y, bins=20)  # Let's try something in between
ax.set_xlabel('Some x values')
ax.set_ylabel('Some y values')
ax.set_title("Expenential Distribution")

# Show the plots
plt.tight_layout()
plt.show()
```

```python
def generate_expon_distribution(size=137, loc=10, scale=20, plot=True):
    #dist_type: norm or expon
    dist = stats.expon(loc=loc, scale=scale)
    fake_y = -dist.rvs(size)
    
    normal_dist = stats.norm(loc=0, scale=5)
    fake_y += normal_dist.rvs(137)
    
    if plot == True:
        fig, axs = plt.subplots(figsize=(20, 15))
        ax = axs
        ax.hist(fake_y, bins=20)  # Let's try something in between
        ax.set_ylim((0, 50))
        ax.set_xlim((-100, 50))
        ax.set_xlabel('Some x values')
        ax.set_ylabel('Some x values')
        ax.set_title("Distribution")
        

generate_expon_distribution()
```
